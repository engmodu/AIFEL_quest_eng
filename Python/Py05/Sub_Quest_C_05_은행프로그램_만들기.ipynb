Q1. Account 클래스: 은행에 가서 계좌를 개설하면 은행이름, 예금주, 계좌번호, 잔액이 설정됩니다. Account 클래스를 생성한 후 생성자(hint: 매직메서드..!!)를 구현해보세요. 생성자에서는 예금주와 초기 잔액만 입력 받습니다. 은행이름은 SC은행으로 계좌번호는 3자리-2자리-6자리 형태로 랜덤하게 생성됩니다. (은행이름: SC은행, 계좌번호: 111-11-111111)

Q2. 클래스 변수: 클래스 변수를 사용해서 Account 클래스로부터 생성된 계좌 객체의 개수를 저장하세요.

Q3. 클래스 변수 출력: Account 클래스로부터 생성된 계좌의 개수를 출력하는 get_account_num() 메서드를 추가하세요.

Q4. 입금 메서드: Account 클래스에 입금을 위한 deposit 메서드를 추가하세요. 입금은 최소 1원 이상만 가능합니다.

Q5. 출금 메서드: Account 클래스에 출금을 위한 withdraw 메서드를 추가하세요. 출금은 계좌의 잔고 이상으로 출금할 수는 없습니다.

Q6. 정보 출력 메서드: Account 인스턴스에 저장된 정보를 출력하는 display_info() 메서드를 추가하세요. 잔고는 세자리마다 쉼표를 출력하세요. (은행이름: SC은행, 예금주: 파이썬, 계좌번호: 111-11-111111, 잔고: 10,000원)

Q7. 이자 지급하기: 입금 횟수가 5회가 될 때 잔고를 기준으로 1%의 이자가 잔고에 추가되도록 코드를 변경해보세요.

Q8. 여러 객체 생성: Account 클래스로부터 3개 이상 인스턴스를 생성하고 생성된 인스턴스를 리스트에 저장해보세요.

Q9. 객체 순회: 반복문을 통해 리스트에 있는 객체를 순회하면서 잔고가 100만원 이상인 고객의 정보만 출력하세요.

Q10. 입출금 내역 기록(이전 이미지 포함): 입금과 출금 내역이 기록되도록 코드를 업데이트하세요. 내역을 출력하는 deposit_history와 withdraw_history 메서드를 추가하세요.

```python
import random

class Account:
    # Q2. 클래스 변수: 계좌 개수 저장
    account_count = 0

    def __init__(self, owner, balance):
        # Q1. 생성자 구현
        self.bank = "SC은행"
        self.owner = owner
        self.balance = balance

        # 계좌번호 랜덤 생성 (3자리-2자리-6자리)
        num1 = random.randint(0, 999)
        num2 = random.randint(0, 99)
        num3 = random.randint(0, 999999)
        self.account_number = f"{num1:03d}-{num2:02d}-{num3:06d}"

        # Q7 & Q10을 위한 추가 변수
        self.deposit_count = 0
        self.deposit_log = []
        self.withdraw_log = []

        # 계좌 생성 시 클래스 변수 증가
        Account.account_count += 1

    # Q3. 계좌 개수 출력 메서드
    @classmethod
    def get_account_num(cls):
        print(f"현재 생성된 계좌 개수: {cls.account_count}개")

    # Q4 & Q7. 입금 메서드 (이자 로직 포함)
    def deposit(self, amount):
        if amount >= 1:
            self.balance += amount
            self.deposit_log.append(amount) # Q10 기록
            self.deposit_count += 1

            # Q7. 5회 입금 시 1% 이자 지급
            if self.deposit_count % 5 == 0:
                interest = int(self.balance * 0.01)
                self.balance += interest
        else:
            print("1원 이상만 입금 가능합니다.")

    # Q5. 출금 메서드
    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance -= amount
            self.withdraw_log.append(amount) # Q10 기록
        else:
            print("잔액이 부족합니다.")

    # Q6. 정보 출력 메서드
    def display_info(self):
        print(f"은행이름: {self.bank}, 예금주: {self.owner}, 계좌번호: {self.account_number}, 잔고: {self.balance:,}원")

    # Q10. 입출금 내역 출력 메서드
    def deposit_history(self):
        print(f"[{self.owner}] 입금 내역: {self.deposit_log}")

    def withdraw_history(self):
        print(f"[{self.owner}] 출금 내역: {self.withdraw_log}")


# Q8. 여러 객체 생성 및 리스트 저장
accounts = []
accounts.append(Account("박지성", 1200000))
accounts.append(Account("차두리", 500000))
accounts.append(Account("이천수", 2500000))
accounts.append(Account("이을용", 10000))
accounts.append(Account("송종국", 2200000))
accounts.append(Account("황선홍", 5300000))
accounts.append(Account("이영표", 200000))
accounts.append(Account("안정환", 1550000))

# Q9. 100만원 이상인 고객 정보만 출력
print("--- 잔액 100만원 이상 고객 목록 ---")
for acc in accounts:
    if acc.balance >= 1000000:
        acc.display_info()

# 동작 테스트 (이자 및 내역 확인)
print("\n--- 동작 테스트 (파이썬 계좌) ---")
py_acc = accounts[3] # '파이썬' 계좌
for i in range(5):
    py_acc.deposit(1000) # 5번 입금 시도 (이자 발생)

py_acc.withdraw(500)
py_acc.display_info()
py_acc.deposit_history()
py_acc.withdraw_history()

# 전체 계좌 개수 확인
print("")
Account.get_account_num()
```
