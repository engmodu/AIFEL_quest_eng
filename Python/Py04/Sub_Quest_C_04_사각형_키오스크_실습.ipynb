```python
menu_temp = ['HOT americano', 'ICE latte', 'ICE mocha', 'ICE choco_latte']
price_temp = [2000, 3000, 3000, 3000]
```

```python
class Kiosk:
    def __init__(self):
        # 메뉴와 가격 설정 (예시 데이터)
        self.menu = ['아메리카노', '카페라떼', '초코라떼']
        self.price = [2000, 3000, 3500]
        self.order_menu = []   # 주문한 메뉴 리스트
        self.order_price = []  # 주문한 메뉴의 가격 리스트
        self.price_sum = 0     # 총 결제 금액

    # 메뉴 출력 메서드
    def menu_print(self):
        print("\n--- 메뉴판 ---")
        for i in range(len(self.menu)):
            print(f"{i + 1}. {self.menu[i]} : {self.price[i]}원")

    # 주문 메서드 (핵심 수정 부분)
    def menu_select(self):
        while True:  # 추가 주문을 위한 무한 루프 시작
            print()
            n = 0
            # 1. 음료 번호 선택
            while n < 1 or n > len(self.menu):
                try:
                    n = int(input("음료 번호를 입력하세요 : "))
                    if 1 <= n <= len(self.menu):
                        break
                    else:
                        print("없는 메뉴입니다. 다시 주문해 주세요.")
                except ValueError:
                    print("숫자를 입력해주세요.")

            # 2. 온도 선택 (문제 2-3 해결 부분)
            t = 0
            while t != 1 and t != 2:
                try:
                    t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                    if t == 1:
                        self.temp = "HOT"
                    elif t == 2:
                        self.temp = "ICE"
                    else:
                        print("1과 2 중 하나를 입력하세요.")
                except ValueError:
                    print("숫자 1 또는 2를 입력해주세요.")

            # 주문 결과 저장 및 출력
            selected_item = self.menu[n-1]
            selected_price = self.price[n-1]

            # 장바구니에 담기
            self.order_menu.append(f"{self.temp} {selected_item}")
            self.order_price.append(selected_price)
            self.price_sum += selected_price

            print(f"-> 주문 추가: {self.temp} {selected_item} ({selected_price}원)")

            # 3. 추가 주문 여부 확인 (핵심 추가 사항)
            print()
            more = input("추가 주문하시겠습니까? (y 입력 시 추가 주문, n 입력 시 종료) : ").lower()
            if more == 'n':
                break
            else:
                self.menu_print() # 다시 메뉴 보여주기

        # 최종 주문 내역 출력
        self.final_order()

    # 최종 결제 정보 출력 메서드
    def final_order(self):
        print("\n" + "="*30)
        print("최종 주문 내역")
        for i in range(len(self.order_menu)):
            print(f"- {self.order_menu[i]} : {self.order_price[i]}원")
        print("-" * 30)
        print(f"총 결제 금액 : {self.price_sum}원")
        print("="*30)

# 실행
a = Kiosk()
a.menu_print()
a.menu_select()
```

```python
# 문제 3-3. 클래스 업데이트
# 앞에서 구현했던 메서드들을 Kiosk 클래스에 추가합니다. 직접 코드를 작성해 보세요!

class Kiosk:
    def __init__(self):
        # 메뉴 및 가격
        self.menu = menu_temp
        self.price = price_temp

        # 상태값 초기화
        self.n = 0
        self.temp = ""
        self.price_sum = 0

    # 메뉴 출력 메서드
    def menu_print(self):
        for i in range(len(self.menu)):
            print(i + 1, self.menu[i], ' : ', self.price[i], '원')

    # 주문 메서드
    def menu_select(self):
        print()
        n = 0

        # 메뉴 선택
        while n < 1 or n > len(self.menu):
            n = int(input("음료 번호를 입력하세요 : "))

            if 1 <= n <= len(self.menu):
                self.n = n
                self.price_sum = self.price[n-1]
            else:
                print("없는 메뉴입니다. 다시 주문해 주세요.")

        # 온도 선택
        t = 0
        while t != 1 and t != 2:
            t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))

            if t == 1:
                self.temp = "HOT"
            elif t == 2:
                self.temp = "ICE"
            else:
                print("1과 2 중 하나를 입력하세요.\n")

        # 주문 확인 출력
        print('주문 음료', self.temp, self.menu[self.n-1], ' : ', self.price[self.n-1], '원')

    # 지불 메서드
    def pay(self):
        print("\n지불할 금액 :", self.price_sum, "원")

        while True:
            pay = input("결제 수단을 선택하세요 (현금: cash 또는 1 / 카드: card 또는 2) : ").lower()

            # 입력값 정규화
            if pay == '1':
                pay = 'cash'
            elif pay == '2':
                pay = 'card'

            if pay == 'cash':
                print("직원을 호출하겠습니다.")
                break
            elif pay == 'card':
                print("IC칩 방향에 맞게 카드를 꽂아주세요.")
                break
            else:
                print("다시 결제를 시도해 주세요.")

    # 주문서 출력
    def table(self):
        print("\n===== 주문표 =====")
        print("메뉴 :", self.temp, self.menu[self.n-1])
        print("가격 :", self.price[self.n-1], "원")
        print("------------------")
        print("총 결제 금액 :", self.price_sum, "원")




a = Kiosk()  # 객체 생성
a.menu_print()  # 메뉴 출력
a.menu_select()  # 주문
a.pay()  # 지불
a.table()  # 주문표 출력
```

```python
# 사각형 넓이를 구하는 클래스
class Square:
    def __init__(self):
        self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'))

        if self.square == 1:
            print('직사각형 함수는 rect()입니다.')

        elif self.square == 2:
            print('평행사변형 함수는 par()입니다.')

        elif self.square == 3:
            print('사다리꼴 함수는 trape()입니다.')

        else:
            print('1, 2, 3 중에서 다시 입력해주세요')

    def rect(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = width * vertical
        result = '직사각형의 넓이는 : ' + str(area)
        return result

    # 평행사변형 메서드
    def par(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = width * vertical

        result = "평행사변형의 넓이는 " + str(area)
        return result

    # 사다리꼴 메서드
    def trape(self):
        top, bottom, height = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = (top + bottom) * height / 2
        result = "사다리꼴의 넓이는 " + str(area)
        return result

a = Square()  # 객체 생성 & 2, 3을 각각 입력해 봅시다.
```
