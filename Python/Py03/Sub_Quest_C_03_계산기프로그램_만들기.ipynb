[QUEST 04] 파이썬을 사용하여 간단한 계산기 프로그램 만들기

사용자로부터 두 개의 정수와 연산자(+, -, *, /, **)를 입력받아 계산 결과를 출력하는 프로그램을 만들어주세요

조건

사용자가 입력한 값이 정수가 아닌 경우 ValueError를 처리하여 적절한 오류 메시지를 출력해야 합니다(함수 -> try)

정수가 입력될 때 까지 "잘못된 입력입니다. 정수를 입력해주세요."를 출력하며 다시 입력받기를 시도합니다.


나눗셈 연산 시 두 번째 정수가 0인 경우 ZeroDivisionError를 처리하여 적절한 오류 메시지를 출력합니다. (try)


사용자가 지원하지 않는 연산자를 입력한 경우 적절한 오류 메시지를 출력합니다.
if esle


math 모듈을 사용하여 제곱 연산(**)을 지원합니다.(import)


계산기를 통해 계속 계산을 할 것인지 입력받습니다. (루프)



```python
import math

# 1. 정수 판별 함수
def is_integer(user_input):
    try:
        int(user_input)
        return True
    except ValueError:
        print("잘못된 입력입니다. 정수를 입력해주세요.")
        return False

# 2. 연산자 판별 및 계산 함수
def calculate(num1, num2, operator):
    if operator == '+':
        return num1 + num2
    elif operator == '-':
        return num1 - num2
    elif operator == '*':
        return num1 * num2
    elif operator == '/':
        try:
            return num1 / num2
        except ZeroDivisionError:
            return "오류: 0으로 나눌 수 없습니다."
    elif operator == '**':
        return math.pow(num1, num2)
    else:
        return "오류: 지원하지 않는 연산자입니다. 정확한 기호를 입력해 주세요."

# 3. 메인 실행 로직
def main():
    while True:
        # 첫 번째 정수 입력
        input_str1 = input("첫 번째 정수를 입력하세요: ")

        # 첫 번째 입력이 정수일 때만 다음으로 진행
        if is_integer(input_str1):
            num1 = int(input_str1)

            # 두 번째 정수 입력
            input_str2 = input("두 번째 정수를 입력하세요: ")

            # 두 번째 입력도 정수일 때만 다음으로 진행
            if is_integer(input_str2):
                num2 = int(input_str2)

                # 연산자 입력 및 계산
                operator = input("연산자를 입력하세요 (+, -, *, /, **): ")
                result = calculate(num1, num2, operator)

                # 결과 출력
                print(f"결과: {num1} {operator} {num2} = {result}")

        # 입력이 틀렸거나 계산이 정상적으로 끝난 후 공통으로 실행됨
        retry = input("\n계속 계산하시겠습니까? 아무키 누르면 계속, 'n'을 누르면 종료): ")
        if retry.lower() == 'n':
            print("계산기 프로그램을 종료합니다.")
            break

if __name__ == "__main__":
    main()
```
